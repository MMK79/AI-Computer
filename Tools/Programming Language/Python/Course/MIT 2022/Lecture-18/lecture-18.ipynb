```python
class Coordinate(object):
    """A coordinate made up of an x and y value"""
    def __init__(self, x, y):
        """Sets x and y values"""
        self.x = x
        self.y = y
    def distance(self, other):
        """Returns euclidian distance between two coordinate objects"""
        x_diff_sq = ( self.x - other.x )**2
        y_diff_sq = ( self.y - other.y )**2
        return (x_diff_sq + y_diff_sq)**0.5
    def to_origin(self):
        self.x = 0
        self.y = 0

# c = Coordinate(3, 4)
# origin = Coordinate(0, 0)
#
# print(f"c's x is {c.x} and origin's x is {origin.x}")
# print(f"c's y is {c.y} and origin's y is {origin.y}")
# print("distance of c from origin is", c.distance(origin))
#
# c.to_origin()
# print("c's x and c's y after use of to_origin method", c.x, c.y)

class Circle(object):
    def __init__(self, center, radius):
        if type(center) != Coordinate:
            raise ValueError("center must be Coordinate") 
        if type(radius) != int:
            raise ValueError("radius must be integer") 
        # assert type(center) == Coordinate, "type of center must be Coordinate"
        # assert type(radius) == int, "type of radius must be integer"
        self.center = center
        self.radius = radius
    def is_inside(self, point):
        assert type(self) == Circle, "Must be Circl"
        assert type(point) == Coordinate, "Point must be Coordinate"
        """Returns True if a point is in self, and False otherwise"""
        return point.distance(self.center) < self.radius
    # same functionality different order, this function and function before are equivilant
    # def is_inside(self, point):
    #     assert type(self) == Circle, "Must be Circl"
    #     assert type(point) == Coordinate, "Point must be Coordinate"
    #     """Returns True if a point is in self, and False otherwise"""
    #     return self.center.distance(point) < self.radius

center = Coordinate(2, 2)
my_circle = Circle(center, 2)
# unit test to raise error for center type
# Circle(1, 2)
# Unit test to raise error for radius type
# my_circle = Circle(center, 'hello')
# p = Coordinate(1, 1) # print True
p = Coordinate(10, 10) # print False
print(my_circle.is_inside(p))
# print(my_circle.is_inside(1))
```

```python
class SimpleFraction(object):
    def __init__(self, n, d):
        assert type(n) == int, "Numerator must be int"
        assert type(d) == int, "Denominator must be int"
        self.num = n
        self.denom = d
    def times(self, other):
        top = self.num * other.num
        bottom = self.denom * other.denom
        return top/bottom 
    def plus(self, other):
        top = self.num*other.denom + other.num*self.denom
        bottom = self.denom*other.denom
        return top/bottom
    def get_inverse(self):
        """Returns a float representing 1/self"""
        top = self.denom
        bottom = self.num
        return top/bottom
        # Other way
        # return self.denom/self.num
    def invert(self):
        """Sets self's num to denom and vice versa
        Returns None."""
        # With use of tuple to change value inside
        (self.num, self.denom) = (self.denom, self.num)
        # Other way
        # SimpleFraction(self.denom, self.num)
        return None

f1 = SimpleFraction(3, 4)
f2 = SimpleFraction(1, 4)
# print(f1.num)
# print(f1.denom)
# print(f1.plus(f2))
# print(f1.times(f2))
print(f"f1 numarator is {f1.num}, and f1 denominator is {f1.denom} before invert method use")
f1.invert()
print(f"f1 numarator is {f1.num}, and f1 denominator is {f1.denom} after invert method use")
print(f"f1 get inverse calculation = {f1.get_inverse()}")
```

```python
class Fraction(object):
    def __init__(self, n, d):
        self.num = n
        self.denom = d
    # print
    def __str__(self):
        return str(self.num) + '/' + str(self.denom)
        # if self.denom == 1:
        #     return str(self.num)
        # else:
        #     return str(self.num) + '/' + str(self.denom)
    # Make * work on our class object
    def __mul__(self, other):
        top = self.num*other.num
        bottom = self.denom*other.denom
        return Fraction(top, bottom)
    # Cast/Convert to float
    def __float__(self):
        return self.num/self.denom
    def reduce(self):
        def gcd(n, d):
            while d != 0:
                (d, n) = ( n%d, d )
            return n
        if self.denom == 0:
            return None
        # Can cause problem cause it is returning a int
        elif self.denom == 1:
            return Fraction( self.num, 1 )
        else:
            greatest_common_divisor = gcd(self.num, self.denom)
            top = int(self.num/greatest_common_divisor)
            bottom = int(self.denom/greatest_common_divisor)
            return Fraction(top, bottom)

f1 = Fraction(3, 4)
f2 = Fraction(2, 4)
f3 = Fraction(5, 1)
# print(f1)
# print(f2)
# print(f3)

# print(f1 * f2)
# behind the scene
# print(f1.__mul__(f2))
# print(Fraction.__mul__(f1, f2))

# print(float(f1))
# print(float(f1 * f2))

c = Fraction(2, 12)
print(c)
print(c.reduce())
print(f1 * f2)
print(Fraction.reduce( f1 * f2 ))
```
