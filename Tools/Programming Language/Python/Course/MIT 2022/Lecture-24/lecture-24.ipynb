```python
import random
import time

def is_sorted(L):
    i = 0
    j = len(L)
    while i + 1 < j:
        if L[i] > L[i + 1]:
            return False
        i += 1
    return True

def bogo_sort(L):
    count = 0
    while not is_sorted(L):
        random.shuffle(L)
        count += 1
    return count

# print("--- BOGO SORT ---")
# L = []
# for i in range(0, 9):
#     L.append(random.randint(0, 100))
# L = [8, 4, 1, 6, 5, 11, 2, 0]
# print('L:       ', L)
# t0 = time.time()
# count = bogo_sort(L)
# print('Sorted L:', L)
# print(count, "shuffles and sorting took: ", time.time() - t0, "s")
```

```python
def bubble_sort(L, detail = False):
    did_swap = True
    while did_swap:
        # Complexity = \Theta(len(L))
        did_swap = False
        for j in range(1, len(L)):
            # Complexity = \Theta(len(L))
            if L[j-1] > L[j]:
                did_swap = True
                L[j], L[j-1] = L[j-1], L[j]
            if detail == True:
                print(L)
        print()
# Complexity = \Theta(len(L**2))

print("--- BUBBLE SORT ---")
L = [8, 4, 1, 6, 5, 11, 2, 0]
print('L:       ', L)
bubble_sort(L, True)
print('Sorted L:', L)
```

```python
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    while (i < len(left)):
        result.append(left[i])
        i += 1
    while (j < len(left)):
        result.append(left[j])
        j += 1
    return result
# Complexity = \Theta(len(left)+len(right)) = \Theta(len(L))

def merge_sort(L, detail = False):
    # Basecase
    if len(L) < 2:
        return L[:]
    # Recursive Step
    else:
        # Devide
        middle = len(L)//2
        left = merge_sort(L[:middle])
        right = merge_sort(L[middle:])
        if detail == True:
            print("Merging", left, "and", right)
        # Conquer
    return merge(left, right)
# Complexity = \Theta(log n)
# Merge-sort overall Complexity = \Theta(nlog n)

print("--- MERGE SORT ---")
L = [8, 4, 1, 6, 5, 11, 2, 0]
print('L:       ', L)
print(merge_sort(L, True))
```
